## Loading the Models

```python
model_1 = pickle.load(open('../input/ubiquant-scikitgb-final/GB_2.sav', 'rb'))
model_2 = pickle.load(open('../input/ubiquant-simple-lgboost-final/simple_LGBM_3.sav', 'rb'))
model_3 = pickle.load(open('../input/ubiquant-simple-xgboost-final/simple_xgb_3.sav', 'rb'))
```

## Loading the scalar

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

# Ensamble the model in the new submission.


```python
import ubiquant
env = ubiquant.make_env()   # initialize the environment
iter_test = env.iter_test()    # an iterator which loops over the test set and sample submission
for test_df, sample_prediction_df in iter_test:
    X = test_df.iloc[:,1:302].values
    X = scaler.fit_transform(X)
    y_pred = np.average(np.array([model_1.predict(scaler.fit_transform(test_df.iloc[:,2:299].values)),model_2.predict(X),model_3.predict(X)]),weights=[0.85,0.05,0.1], axis=0)
    sample_prediction_df.iloc[:,1] = y_pred
    env.predict(sample_prediction_df)   # register your predictions
```
