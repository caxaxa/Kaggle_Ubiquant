## 1. Import libraries 

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import xgboost as xgb
import pickle
import gc
```

## 2. Download datasets 

```python
# Generate a dictionary to read the csv using float32

def Convert(a):
    it = iter(a)
    res_dct = dict(zip(it, it))
    return res_dct



feature_list = []
for i in range(300):
    feature_list.append('f_{}'.format(i))
    feature_list.append('float32') # comment this line to get a list with the name of the features
        
feature_dict = Convert(feature_list)
```

### Opening the one million additional data

```python
df_train = pd.read_csv('../input/ubiquant-market-prediction/train.csv', dtype = feature_dict, skiprows = lambda x: x in range(1,2000000), header = 0 )

df_train.head()
```

```python
df_train = df_train.drop('row_id', axis = 1)
df_train.tail()
```

## Continue the training

1 - Load the model

2 - Prepare New Data
 use warm_start = True
 
3 - Fit the model

4- Save new model

```python
model = pickle.load(open('../input/ubiquant-simple-lgboost-2/simple_xgb_2.sav', 'rb'))

```

```python
%%time
# model 2 - Use this model for the unobservable investments
X2 = df_train.drop(['target','time_id'], axis=1)
y2 = df_train['target']

del df_train
gc.collect()
```

```python
X2 = X2.values
y2 = y2.values

gc.collect()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X2[:,1:400]= scaler.fit_transform(X2[:,1:400])

X2[:,0] = X2[:,0].astype('int16')
```

```python
%%time

model = model.fit(X2, y2)
```

```python
model.score
```
